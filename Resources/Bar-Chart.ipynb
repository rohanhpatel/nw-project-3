```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('haunted_places.csv')
df.head()
```

```python
df.info()
```

```python
top_10_state = df['state'].value_counts().head(10)

top_10_state
```

```python
plt.figure(figsize=(15, 6))
bars = plt.bar(top_10_state.index, top_10_state.values, color = 'lightblue')

# Add data labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom')

plt.xlabel('State')
plt.ylabel('# of Haunted Places')
plt.title('Top 10 Most Haunted States in USA')
plt.xticks(rotation=45)
plt.show()
```
